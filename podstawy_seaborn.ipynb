**Cwiczenie **

Narysuj wykres countplot gdzie na osi x bedzie plec a jako kolor słupków uzyty zostanie level.
Sprobuj zmienic kolejnosc Mezczyzn i Kobiet (parametr order) oraz leveli (parametr hue_order)
Dostosuj tytul, kolory, legende - wedlug uznania.

```python
g = sns.countplot(data=df, x='sex', hue='level', hue_order=['senior', 'mid', 'junior'] , stat='percent', order=['M', 'F'])
g.set_xticks([0, 1])  # pozycje ticków
g.set_xticklabels(['Mężczyźni', 'Kobiety'])
```

**Cwiczenie ** 

Podobny wykres mozemy narysowac uzywajac histplot:
https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot 

Sprobuj przerobic wykres z cwiczenia pierwszego uzywajac histplot
zwroc uwage na parametry:

    - stat - jak zmienia sie wykrtes zaleznosci od wartosci parametru?
    - discrete, 
    - multiple - jak zmienia sie wykrtes zaleznosci od wartosci parametru? {'layer', 'dodge', 'stack', 'fill'}
    - shrink.

Czy uda sie osiagnac taki sam wykres jak wyzej?

```python
g = sns.histplot(data=df, x='sex', hue='level', discrete=True, shrink=0.8, multiple='dodge')
```

```python
## ten wykres pokazuje jak rozklada sie poziom stanowisk dla kazdej z plci

g = sns.histplot(data=df, x='sex', hue='level', discrete=True, shrink=0.8, multiple='fill')
for ax_container in g.containers:
    g.bar_label(ax_container, label_type='center', fmt='{:.1%}')
sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1))
```

## Histogram - ciag dalszy

```python
### Rozklad wyplaty:
sns.histplot(data=df, x='salary')

# binwidth=100 - szerokosc na 100 
# bin = 30 - dzieli na 30 przedzialow
# bins = [0, 1000, 2000, 3000]
```

```python
sns.histplot(data=df, x='salary', hue='level', binwidth=100) 
## tu dosyc mocno nakladaja sie te wykresy i nie widac roznicy w rozkladach... 

```

```python
## Rozwiazania:
## inny "element" niz slupki
sns.histplot(data=df, x='salary', hue='level', binwidth=100, element='step') 
plt.show()
sns.histplot(data=df, x='salary', hue='level', binwidth=100, element='poly') 
plt.show()

```

```python
# 2. osobny wykres dla kazdego levela
# https://seaborn.pydata.org/generated/seaborn.FacetGrid.html

## sns.FacetGrid(data, row, columns) pozwala na zdefiniowane siatki wykresow
## sns.map(<obiejkt wykresu>, <parametry wykresu>) pozwala na narysowanie wykresu

g = sns.FacetGrid(data= df, row='level', aspect = 3) ## szerokosc wykresu to hight * aspect
g.map(sns.histplot, 'salary', bins=30)

## do facetGrid mozna dodac , hue='level' aby uzyska kolory dla kazdego levelu
```

**Cwiczenie**

Narysuj analogiczny wykres z podzialem po plci (row='sex'); Niech level bedzie kolorem elementow (hue='level')

```python
g = sns.FacetGrid(data= df, row='sex', hue='level', aspect = 3) ## szerokosc wykresu to hight * aspect
g.map(sns.histplot, 'salary', binwidth=100, element='poly')
g.figure.suptitle('Rozkład wynagrodzeń', fontsize=16)
g.set_titles(row_template='Płeć: {row_name}', size=12, weight='bold')
plt.tight_layout()
```
