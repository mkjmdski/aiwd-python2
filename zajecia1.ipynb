### Zmienne

W języku Python istnieją dwa główne typy zmiennych:

- mutable (zmienne mutowalne lub zmienialne) – czyli takie, których zawartość można zmieniać po utworzeniu obiektu:
    - Listy (`[...]`- uporządkowana, mutowalna sekwencja elementów, która może przechowywać wartości różnych typów)
    - Dictionary (`{key: value}`; key musi być niemutowalnym obiektem; value - dowolnym)
    - pandas.DataFrame
- immutable (zmienne niemutowalne lub niezmienialne) – czyli takie, których zawartości nie można zmienić po utworzeniu obiektu.
    - tuple (`(...)` - uporządkowana, niemutowalna sekwencja elementów, która może przechowywać wartości różnych typów.)
    - stringi

Nie należy używać mutowalnych obiektów (np. list, słowników) jako wartości domyślnych parametrów funkcji, ponieważ są one współdzielone między wszystkimi wywołaniami tej funkcji, co może prowadzić do nieoczekiwanych błędów.

````python
## Przyklad 

def add_element(elt: int, p_list: list[int])-> list:
    """Add new element to list; return new list 
    (is this realy new list?? )
    """
    p_list.append(elt)
    return p_list

A = [1,2,3]
print(A)

C = add_element(4, A)
print('A:', A)
print('C:', C)

D = add_element(5, A)
print('A:', A)
print('C:', C)
print('D:', D)
````

### Python - jakość kodu  

https://realpython.com/python-code-quality/ 

Istnieją  standardy tego jak kod powinine wygladac. 
https://peps.python.org/pep-0008/ - standard dla Pythona.
czytelność kodu jest wazna - i dla nas i dla innych. 

Najwazniejsze wytyczne:
- wszystkie importy na górze modulu / notebooka
- sensowne nazwy zmiennych, funckji i klas 
- docstring - umozliwiaja documentowanie funkcji i modułow w kodzie
- typehint - python nie wymaga deklaracji zmiennych, ale dobrze jest dodawac typehinty w deklaracjach funckji.

````python
# przyklad: low quality:
def mul(a, b):
     return a * b

# przyklad: high quality:
def multiply(a: float, b: float = 2) -> float:
     """Multiply two numbers.
     Args:
         a (float): First number.
         b (float): Second number.
     Returns:
         float: Product of a and b.
     """
     return a * b

print(multiply(4))
print(multiply(4, 5))

## uwaga: typehinty sa tylko informacyjne, w zaden sposob nie wymuszają typu. 
multiply('a')
````
